```python
import pandas as pd
```

```python
nyc=pd.read_csv("/content/ave_hi_nyc_jan_1895-2018.csv", header='infer')
nyc
```

```python
nyc.columns
```

```python
nyc.colums=['Date','Temperature','Anomaly']
nyc.head(6)
```

```python
nyc.Date=nyc.Date.floordiv(100)
nyc.head()
```

#Temperature=m*Date+c                       

```python
from sklearn.model_selection import train_test_split
```

```python
nyc['Date'].values
```

```python
X = nyc[['Date']].values  # Use double brackets to ensure 'Date' is treated as a DataFrame
y = nyc['Value'].values

# Split the data
X_train, X_test, y_train, y_test = train_test_split(nyc.Date.values.reshape(-1, 1), nyc.Value.values.reshape(-1,1), random_state=104)
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regression = LinearRegression()
linear_regression
```

```python
linear_regression.fit(X_train, y_train)
```

```python
linear_regression.coef_
```

```python
#Value=0.83039216 -21.385294117647042
```

```python
linear_regression.intercept_
```

```python
0.70710814*2098- 21.385294117647042
```

```python
predicted = linear_regression.predict(X_test)
predicted
```

```python
predicted=(lambda date: linear_regression.coef_*date+linear_regression.intercept_)(2030)
predicted
```
